### SED
http://www.jrdb.com/program/Sed/sed_doc.txt

### SRB
http://www.jrdb.com/program/Srb/srb_doc.txt

```python
data_dir = 'data/SED_2022'

# get list of file names in directory
file_names = os.listdir(data_dir)

# filter file names to those starting with 'SED' and ending with '.txt'
sed_file_names = [f for f in file_names if f.startswith('SED') and f.endswith('.txt')]

# load text data and parse into data frames
dfs = []
for file_name in tqdm(sed_file_names):
    loader = load.FileLoader()
    text_data = loader.load(os.path.join(data_dir, file_name))
    parser = parse.JrdbDataParser()
    df = parser.parse(text_data, 'SED', is_japanese=True)
    dfs.append(df)

# concatenate data frames into one
SED_df = pd.concat(dfs, ignore_index=True)
```

```python
SED_df = pd.read_csv('train_data/SED_2022.csv')
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.utils import resample

# 目的変数の作成
# 3着以内を1、それ以外を0とする
# '着順'カラムを使用

SED_df['target'] = SED_df['着順'].apply(lambda x: 1 if x <= 3 else 0)

categorical_columns = ['レースキー', '日', '年月日', '馬名', '条件', 'グレード', 'レース名', 'レース名略称', '騎手名', '調教師名', 'レースペース', '馬ペース', '1(2)着馬名']

# 各カテゴリカル列を文字列型に変換し、ラベルエンコーディングを適用
SED_df[categorical_columns] = SED_df[categorical_columns].apply(lambda col: LabelEncoder().fit_transform(col.astype(str)))

# データの不均衡を解消
# 3着以内のデータ（クラス1）とそれ以外のデータ（クラス0）を分ける
class_1 = SED_df[SED_df['target'] == 1]
class_0 = SED_df[SED_df['target'] == 0]

# クラスのサンプル数を均等にする
# ここでは多数クラスのサンプル数を少数クラスに合わせてダウンサンプリング
class_0_downsampled = resample(class_0, replace=False, n_samples=len(class_1), random_state=123)

# ダウンサンプリングされたデータを結合
SED_df= pd.concat([class_1, class_0_downsampled])

# 日付でソートし、最新の20%をテストデータとする
# '年月日'カラムを使用
SED_df['年月日'] = pd.to_datetime(SED_df['年月日'])
SED_df = SED_df.sort_values(by='年月日')
train_SED_df = SED_df.iloc[:int(len(SED_df) * 0.8)]
test_SED_df = SED_df.iloc[int(len(SED_df) * 0.8):]

# 不要なカラムの削除
# '着順'と'年月日'カラムを削除
train_SED_df = train_SED_df.drop(['着順', '年月日'], axis=1)
test_SED_df = test_SED_df.drop(['着順', '年月日'], axis=1)

# 学習データとテストデータの確認
print('Training SED_df shape:', train_SED_df.shape)
print('Test SED_df shape:', test_SED_df.shape)
```

```python
# カラムのデータ型を確認
print(train_SED_df.dtypes)
```

```python
# カテゴリカル変数の特定

categorical_cols = train_SED_df.select_dtypes(include=['object']).columns

# テキストデータの特定
# ここでは、テキストデータがどのカラムに含まれているかをユーザーからの指示に基づいて判断します。

# 数値変数の特定
numeric_cols = train_SED_df.select_dtypes(include=['int64', 'float64']).columns

# カテゴリカル変数のワンホットエンコーディング
train_SED_df = pd.get_dummies(train_SED_df, columns=categorical_cols)
test_SED_df = pd.get_dummies(test_SED_df, columns=categorical_cols)


# 数値変数の標準化（必要に応じて）
# ここでは、標準化の必要性をユーザーからの指示に基づいて判断します。

# 前処理後のデータの確認
print('Training data shape after preprocessing:', train_SED_df.shape)
print('Test data shape after preprocessing:', test_SED_df.shape)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.metrics import f1_score, roc_auc_score

# データの分割
X_train, X_test, y_train, y_test = train_test_split(train_SED_df.drop(['target'], axis=1), train_SED_df['target'], test_size=0.2, random_state=42)


# LightGBMのパラメータ設定
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'binary_logloss',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'max_depth': -1,
    'min_child_samples': 20,
    'max_bin': 255,
    'verbose': -1
}

# モデルのトレーニング
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)

gbm = lgb.train(params, lgb_train, num_boost_round=500, valid_sets=lgb_eval, early_stopping_rounds=20)

# モデルの評価
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
# 予測値を二値に変換
y_pred_binary = (y_pred >= 0.5).astype(int)
f1 = f1_score(y_test, y_pred_binary)
auc = roc_auc_score(y_test, y_pred)

# 評価指標の表示
print(f'F1 Score: {f1}')
print(f'AUC Score: {auc}')
```
